```python
data = pdr.get_data_yahoo("AAPL", start="2017-01-01", end="2017-04-30")
```

```python
df=pd.DataFrame(data)
```

```python
df.tail()
```

```python
mean_values=[round(df['Open'].mean()),round(df['Low'].mean()),round(df['Volume'].mean())//10000000,round(df['Adj Close'].mean())]
mean_values
# Volume in 1 million i.e 10 lakhs
```

```python
# plt.pie(mean_values)
labels = ['Open', 'Low', 'Volume', 'Adj Close']
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(mean_values, labels=labels, autopct='%.1f%%',
       wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       textprops={'size': 'x-large'})
ax.set_title('Apple Stock Mean Values')
plt.tight_layout()
```
